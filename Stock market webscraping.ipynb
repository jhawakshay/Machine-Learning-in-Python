```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

#Creating an empty list
company = []


########Creating a empty dataframe
#df = pd.DataFrame(columns=range(1,14))

with open("E:/Stock Market/2020/2020-02-14-NSE-EQ.txt") as fo:
# Create an empty list
    for rec in fo:
        rec1 = rec.split(',')[0]
        company.append(rec1)
del(company[0])
columns1 = ["stock", "metric","value"]
df       = pd.DataFrame(columns = columns1, index = range(0,59))
df1      = pd.DataFrame()

######### Pulling information ########
for i in range(0,len(company)):   
  stock = company[i] + '.NS'
  path1 = "https://in.finance.yahoo.com/quote/"
  path2 = "/key-statistics?p="
  url = path1 + str(stock) + path2 + str(stock)
  response = requests.get(url)
  soup = BeautifulSoup(response.text, 'lxml')  
  row_marker = 0

  for table in soup.find_all('table'):
    for row in table.find_all('tr'):
        column_marker = 1
        columns = row.find_all('td')
        for column in columns:
            df.iloc[row_marker,0] = company[i]
            df.iloc[row_marker,column_marker] = column.get_text()
            column_marker += 1
        row_marker += 1
     
  df1 = df1.append(df,ignore_index=True)  
```
